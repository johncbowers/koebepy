```python
from koebe.geometries.orientedProjective2 import *
from koebe.geometries.euclidean2 import *

from koebe.graphics.euclidean2viewer import E2Viewer, makeStyle

import random
```

```python
rad = 100
depth = 7

generators = [
    DiskOP2.fromCenterAndRadius(PointOP2(-100, -100), rad), 
    DiskOP2.fromCenterAndRadius(PointOP2(-100, 100), rad), 
    DiskOP2.fromCenterAndRadius(PointOP2(100, -100), rad),
    DiskOP2.fromCenterAndRadius(PointOP2(100, 100), rad)
]

def invertDisks(disks, generators):
    return set([dp.invertThrough(d) 
            for dp in disks 
            for d in generators 
            if ((not d == dp) and (4.0*dp.a*dp.a) > 0 and dp.radiusSq > 0)
           ])

ds = set(generators)
for _ in range(depth):
    ds = ds | invertDisks(ds, generators)


print(f"Generated {len(ds)} circles.")
```

```python
viewer = E2Viewer(600, 600, 1)
drawable = [d for d in ds if (4.0*d.a*d.a) > 0 and d.radiusSq > 0]
viewer.addAll(drawable)
blackSty = makeStyle(stroke="rgba(0,0,0,255)", strokeWeight=0.2)
redSty = makeStyle(stroke="rgb(255, 0, 0)", strokeWeight=0.5)
blueSty = makeStyle(stroke="rgb(0, 0, 255)", strokeWeight=0.5)
for d in drawable: 
    if random.random() < 0.03:
        viewer.setStyle(d, redSty)
    elif random.random() > 0.97:
        viewer.setStyle(d, blueSty)
    else:
        viewer.setStyle(d, blackSty)
viewer.show()
```
